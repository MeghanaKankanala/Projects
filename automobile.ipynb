# creating simple linear regression model using the best feature

```python
x=car_data[["enginesize"]]

y=car_data["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.7,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model=LinearRegression()
model.fit(x_train,y_train)
model.intercept_ ,model.coef_[0]


```

```python
y_pred=model.predict(x_test)
print(r2_score(y_test,y_pred)) # r2 value using one feature
```

## now considering features whose correlation value >0.7 w.r.t target vector (price) and building multi regression model using recursive feature elimination 

```python
X=pd.DataFrame()
X["enginesize"]=car_data["enginesize"]
X["horsepower"]=car_data["horsepower"]
X["curbweight"]=car_data["curbweight"]
X["carwidth"]=car_data["carwidth"]
X["citympg"]=car_data["citympg"]
X["highwaympg"]=car_data["highwaympg"]
X.head()
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
estimator = SVR(kernel="linear")
rfe = RFE(estimator, n_features_to_select=3, step=1)
rfe.fit(X,y)
X=rfe.transform(X)
X=pd.DataFrame(X, columns=["enginesize","horsepower","carwidth"])
X
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,y,train_size=0.7,random_state=1)
model.fit(x_train,y_train)
print(model.intercept_ ,model.coef_)
y_pred=model.predict(x_test)
print(r2_score(y_test,y_pred))
```

#### Got 0.8012049630690673 R2 score by reducing  features to 3("enginesize","horsepower","carwidth") which means this model explains that 80% the variance in y is acountable by  enginesize,horsepower,carwidth
these are the factors that drive prize and have positive coefficients which means these are directly proportional to the target vector prize
